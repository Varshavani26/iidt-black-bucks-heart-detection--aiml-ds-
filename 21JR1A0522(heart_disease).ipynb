```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/heart-disease.csv"
heart_data = pd.read_csv(file_path)

# Display the first few rows
print(heart_data.head())

# Check for missing values
missing_values_count = heart_data.isnull().sum()
print(missing_values_count)

# Convert categorical variables into dummy/indicator variables
heart_data_processed = pd.get_dummies(heart_data, columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'], drop_first=True)

# Display the first few rows of the processed dataset
print(heart_data_processed.head())

# Average age of patients with heart disease
avg_age_with_disease = heart_data[heart_data['target'] == 1]['age'].mean()

# Average age of patients without heart disease
avg_age_without_disease = heart_data[heart_data['target'] == 0]['age'].mean()

print("Average age of patients with heart disease:", avg_age_with_disease)
print("Average age of patients without heart disease:", avg_age_without_disease)

# Distribution of chest pain types
chest_pain_type_distribution = heart_data['cp'].value_counts()
print(chest_pain_type_distribution)

# Correlation between thalach and age
thalach_age_correlation = heart_data['thalach'].corr(heart_data['age'])
print("Correlation between thalach and age:", thalach_age_correlation)

# Effect of sex on the presence of heart disease
sex_effect_on_disease = heart_data.groupby('sex')['target'].mean()
print(sex_effect_on_disease)

# Histogram of age distribution
plt.hist(heart_data['age'], bins=20, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Patients')
plt.show()

# Bar chart of chest pain types
chest_pain_type_distribution.plot(kind='bar')
plt.xlabel('Chest Pain Type')
plt.ylabel('Frequency')
plt.title('Distribution of Chest Pain Types')
plt.show()

# Scatter plot of thalach vs age
plt.scatter(heart_data['age'], heart_data['thalach'])
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.title('Relationship between Age and Maximum Heart Rate')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
url="/content/heart-disease.csv"
data = pd.read_csv(url)

# Displaying the first few rows of the dataset
data.head()
```

```python
# Let's check for missing values first
data.isnull().sum()
```

```python
# If there are missing values, handle them appropriately, for example by filling with mean or median
# data.fillna(data.mean(), inplace=True)  # Example

# Convert categorical variables into dummy/indicator variables
data = pd.get_dummies(data, drop_first=True)

# Display the first few rows after preprocessing
data.head()
```

```python
# Calculate the average age of patients with and without heart disease
average_age_with_disease = data[data['target'] == 1]['age'].mean()
average_age_without_disease = data[data['target'] == 0]['age'].mean()
print(f"Average age of patients with heart disease: {average_age_with_disease}")
print(f"Average age of patients without heart disease: {average_age_without_disease}")

# Determine the distribution of chest pain types among patients
chest_pain_distribution = data['cp'].value_counts()
print(f"Chest pain type distribution:\n{chest_pain_distribution}")

# Find the correlation between thalach (maximum heart rate) and age
correlation_thalach_age = data['thalach'].corr(data['age'])
print(f"Correlation between maximum heart rate (thalach) and age: {correlation_thalach_age}")

# Analyze the effect of sex on the presence of heart disease
sex_heart_disease = data.groupby('sex')['target'].mean()
print(f"Effect of sex on heart disease:\n{sex_heart_disease}")
```

```python
# Plot a histogram of the age distribution of patients
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar chart showing the distribution of chest pain types among patients
plt.figure(figsize=(10, 6))
chest_pain_distribution.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot a scatter plot to show the relationship between thalach (maximum heart rate) and age
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['thalach'], color='purple', edgecolor='black')
plt.title('Relationship between Age and Thalach')
plt.xlabel('Age')
plt.ylabel('Thalach (Maximum Heart Rate)')
plt.show()
```

```python
## Create a box plot to compare the age distribution of patients with and without heart disease
data.boxplot(column='age', by='target')
plt.title('Age Distribution by Heart Disease Status')
plt.suptitle('')
plt.xlabel('Heart Disease Status (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()
```

```python
# Correlation matrix for all numerical features
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
# Rolling mean analysis on cholesterol levels
data['chol_rolling_mean'] = data['chol'].rolling(window=5).mean()

plt.figure(figsize=(10, 6))
plt.plot(data['chol'], label='Cholesterol')
plt.plot(data['chol_rolling_mean'], label='Rolling Mean (window=5)', color='red')
plt.title('Rolling Mean Analysis of Cholesterol Levels')
plt.xlabel('Index')
plt.ylabel('Cholesterol')
plt.legend()
plt.show()
```
